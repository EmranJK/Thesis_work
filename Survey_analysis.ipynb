```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = '/content/sample_data/survey results.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# [1] [2] [3]

fig, ax =plt.subplots(1,3)
sns.countplot(data=df, x="Gender", ax=ax[0])
sns.countplot(data=df, x="Field of Study", ax=ax[1])
sns.countplot(data=df, x="Age Range", ax=ax[2])
fig.set_size_inches(20.5, 5.5)
fig.show()
```
